```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
import warnings
warnings.filterwarnings('ignore')
```

Data Collection & Processing


```python
#loading dataset from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
#loading the data into pandas library
data_frame=pd.DataFrame(breast_cancer_dataset.data,columns=breast_cancer_dataset.feature_names)
#printing the first five rows
data_frame.head()
```

```python
#adding a target column for 1 and 0 for the given dataset
data_frame['label']=breast_cancer_dataset.target
#printing the last five rows of the dataframe
data_frame.tail()
```

Separating the features and target

```python
X=data_frame.drop(columns=['label'],axis=1)
Y=data_frame['label']
print(X)
```

```python
print(Y)
```

```python
X_train, X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=2)
print(X.shape,X_train.shape,X_test.shape)
#standardise the data for higher acccuracy and this shd be done after spitting the data
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_std=sc.fit_transform(X_train)
X_test_std=sc.transform(X_test)
```

Building a Newral network

```python
#importing tensorFlow and Keras
import tensorflow as tf
tf.random.set_seed(3)#random way of generating values
from tensorflow import keras#tensorflow uses keras as the backend
```

```python
#setting up the layers of my neural network
#Input layer – takes raw data (e.g., pixels of an image).
#Hidden layers – process the data through mathematical operations.
#Output layer – gives the final result (e.g., “this is a cat”).
model=keras.Sequential ([
    keras.layers.Flatten(input_shape=(30,)),#to convert the dataset of 30 features to a single dimensional array
    keras.layers.Dense(20,activation='relu'),#hidden layer relu staands for rectified linear units
    keras.layers.Dense(2,activation='sigmoid')#output layer and the number of inputs are equal to the number of output that can be set(0 or1)
])
```

```python
#compailing the neural network
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
#training the neural network
history=model.fit(X_train_std,Y_train,validation_split=0.1,epochs=10)#epochs says how many times the model shd go through the data
```

visualizing the accuracy and the loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data','validation data'],loc='lower right')
plt.show()
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('Loss')
plt.xlabel('epoch')
plt.legend(['training data','validation data'],loc='upper right')
plt.show()
```

```python
#Accuracy of the model on the test data
loss,accuracy=model.evaluate(X_test_std,Y_test)#comaparing the testing label with orginal label
print('Test Accuracy:',accuracy)
```
